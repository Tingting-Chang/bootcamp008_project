```python
import re
import pandas as pd
```

```python
concerts = pd.read_csv("/home/andrew/bootcamp008_project/Project2-WebScraping/AndrewFeierman/localvenues/concerts.csv")
```

```python
concerts.head()
```

```python
concerts['location'] = pd.Series()
for url in range(len(concerts['url'])):
    concerts['location'][url] = re.search("(us-)([^?]*)", concerts['url'][url]).group(2)
```

```python
concerts
```

```python
concerts.to_csv("concerts_clean.csv")
```

```python
concerts['time'] = pd.Series()
for date in range(len(concerts['date'])):
    if "T" in concerts['date'][date]:
        concerts['time'][date] = re.search('(201.[^T"]*)T?([^"]*)', concerts['date'][date]).group(2)
        concerts['date'][date] = re.search('(201.[^T"]*)T?([^"]*)', concerts['date'][date]).group(1)
    else:
        concerts['time'][date] = "N/A"
        concerts['date'][date] = re.search('(201.[^T"]*)T?([^"]*)', concerts['date'][date]).group(1)
```

```python
concerts.to_csv("concerts_clean.csv")
```
