```python
# Concatenate, Order Col and Export to data.json
d1=pd.read_json('trans_01.json')
d2=pd.read_json('trans_02.json')
d3=pd.read_json('trans_03.json')
d4=pd.read_json('trans_04.json')
d5=pd.read_json('trans_05.json')
d6=pd.read_json('trans_06.json')
d7=pd.read_json('trans_07.json')
d8=pd.read_json('trans_08.json')
d9=pd.read_json('trans_09.json')
d10=pd.read_json('trans_10.json')

frames = [d1, d2, d3, d4, d5, d6, d7, d8, d9, d10]
data = pd.concat(frames)
data.columns
data = data[['Date','Year','Month','Period','Name','Target','Text']] 
data['index'] = range(0,len(data),1)
data.set_index('index',inplace=True)

data.to_json('data.json')
```
